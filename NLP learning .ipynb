# Stop Words

```python
nltk.download('stopwords')
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
sentences = sent_tokenize(paragraph)
```

```python
# Set up stopwords
stop_words = set(stopwords.words('english'))
```

```python
for i, sentence in enumerate(sentences, start=1):
    cleaned_sentence = re.sub(r'\W+', ' ', sentence)  # Replace non-word characters with spaces
    cleaned_sentence = re.sub(r'\s+', ' ', cleaned_sentence).strip()  # Remove extra white spaces
    
    # Tokenize the cleaned sentence
    words = word_tokenize(cleaned_sentence)
    
    # Remove stopwords
    filtered_words = [word for word in words if word.lower() not in stop_words]
    
    cleaned_sentence = ' '.join(filtered_words)
    
    print(f"Sentence {i}: {cleaned_sentence}\n")
```

# Word cloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Create a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(sentences))

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
